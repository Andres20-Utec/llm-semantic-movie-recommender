```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

movies = pd.read_csv("movies_cleaned.csv")
```

```python
movies
```

```python
movies["tagged_overview"].to_csv(
    "tagget_description.txt",
    index=False,
    sep="\n",
    header=False
)
```

```python
raw_documents = TextLoader(r"tagget_description.txt", encoding='utf-8').load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_movies = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings()
)
```

```python
query = "A movie about a war between humans and robots"
docs = db_movies.similarity_search(query, k=10)
docs
```

```python
movies[movies['id']== int(docs[0].page_content.split(" ")[0].strip())]
```

```python
def retriever_semantic_recommendations(query: str, top_k: int = 10):
    recs = db_movies.similarity_search(query, k=50)
    
    movies_list = []
    
    for i in range(len(recs)):
        movies_list.append(int(recs[i].page_content.strip('"').split(" ")[0]))
        
    return movies.loc[movies['id'].isin(movies_list)].head(top_k)
```

```python
retriever_semantic_recommendations("A movie where a dog waits for his owner to come back home", top_k=10)
```
